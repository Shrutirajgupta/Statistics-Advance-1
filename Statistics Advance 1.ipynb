1. What is a random variable in probability theory?

 A random variable is a numerical outcome of a random experiment. It assigns numbers to the outcomes of a random process, such as rolling a die or flipping a coin.


2. What are the types of random variables?

 Discrete Random Variables: Take specific, countable values (e.g., the number of heads in coin tosses).

 Continuous Random Variables: Take any value within a range (e.g., temperature readings).


3. What is the difference between discrete and continuous distributions?

 Discrete distributions: Represent probabilities for discrete random variables, such as rolling a die. Probabilities are assigned to individual outcomes.

 Continuous distributions: Represent probabilities for continuous random variables. Probabilities are calculated over intervals using a Probability Density Function (PDF).


4. What are probability distribution functions (PDF)?

 A Probability Distribution Function (PDF) describes the likelihood of a random variable taking on a specific value. For discrete variables, it gives probabilities for each outcome; for continuous variables, it describes the density of probabilities across an interval.


5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

 PDF provides the probability of a specific value (for discrete) or the density (for continuous).

 CDF gives the cumulative probability of the random variable being less than or equal to a given value.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

 The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's original distribution. It is critical because it allows statisticians to use normal distribution properties for inferential statistics.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(p, x):
    return p if x == 1 else 1 - p

p = 0.7  # Probability of success
x = 1  # Value (1 for success, 0 for failure)
print("Bernoulli PDF:", bernoulli_pdf(p, x))

```

17. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

# Sample data
data = [12, 15, 14, 10, 13, 16, 12, 14, 11]
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)

# Confidence interval (95%)
z_critical = norm.ppf(0.975)  # Two-tailed
margin_of_error = z_critical * (sample_std / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Confidence Interval:", confidence_interval)

```

18. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy.stats import norm

# Generate data
data = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, Std=15

# Confidence interval (95%)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

z_critical = norm.ppf(0.975)  # Two-tailed
margin_of_error = z_critical * (sample_std / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Sample Mean:", sample_mean)
print("Confidence Interval:", confidence_interval)

```

22. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

23. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np

# Generate data from a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std Dev=10

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate IQR
q1 = np.percentile(data, 25)  # 25th percentile
q3 = np.percentile(data, 75)  # 75th percentile
iqr = q3 - q1

print("Range:", data_range)
print("Interquartile Range (IQR):", iqr)


```

24. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import matplotlib.pyplot as plt

# Generate a dataset
data = np.random.normal(loc=100, scale=15, size=1000)  # Mean=100, Std Dev=15

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Visualization: Original vs Normalized
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', label='Original Data')
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, alpha=0.7, color='green', label='Z-score Normalized Data')
plt.title("Z-score Normalized Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```
